### Due to the high dimensionality of this dataset, it is very important to use PCA (Principal Component Analysis) to reduce dimensionality, make predictions more accurate, and facilitate the visualization of graphs. However, before using the PCA technique, it is necessary to prepare  the code, because in this dataset with many outliers and an asymmetrical distribution, this greatly reduces the efficiency of the algorithm. Therefore, the data will be standardized using the 'StandardScaler' function. This ensures that the PCA works efficiently and generates good insights about the dataset.

```python
# Standardize features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Before Standardize
print(X)

# After Standardize
print(X_scaled)
```

```python
# Import PCA library

from sklearn.decomposition import PCA
```

```python
# The dataset already was splitted in X and y above, in array format.
# Now, with the help of a graph, I will visualize the ideal number of features so that the variance is 99.5%.
# This plot shows PCA with no normalized dataset

spam_or_not_no_label = df.drop(df["is_spam"])
pca = PCA().fit(spam_or_not_no_label)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.grid(True)
plt.title('PCA No-Normalized')
plt.show()
```

```python
# This plot shows PCA with normalized dataset

spam_or_not_no_label = X_scaled
pca = PCA().fit(spam_or_not_no_label)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.grid(True)
plt.title('PCA Normalized')
plt.show()
```

##### In these two graphs, there is a significant difference, despite them appearing to be the same X (dataset with only independent columns). However, from the moment I normalized my data using StandardScaler(), I ensured that the standard deviation did not exceed 1.


##### This is evident when we see in the graph that, with this normalization, I need 50 variables to reach this variance value(99.5%), showing that, in fact, the dataset has high dimensionality. Therefore, the reduction seemed small when comparing 57 features, reducing to 50.


##### In the PCA plot without normalization, 2 features are already sufficient to reach the 99.5% variance mark. Although this low number seems advantageous, care must be taken, because starting from this principle, there is a risk of ignoring important variables.

#### Explain in your own words what the “Curse of Dimensionality” is and how it may affect your analysis going forward in the context of this problem.

- **Curse of Dimensionality** refers to the challenges that arise when analyzing data with a large number of features (or dimensions). In simple terms, when we have more dimensions, it becomes harder for algorithms to find patterns and make accurate predictions.

- When we think about the curse of dimensionality, we can think of it this way: when a machine uses an algorithm to make predictions or classifications, it understands patterns. These patterns are influenced by dimensionality. Dimensionality is not the same as a feature, but typically, the number of features in the dataset corresponds to the number of dimensions.

- **Example with a Triangle:**
    - Imagine a triangle. If I ask you to draw a triangle with three lines, that's a simple task (Note: this can be any type of triangle, such as isosceles or equilateral).
    - For a dataset with 3 dimensions, it becomes easier and more accurate to find a pattern, and less data is required.
    - However, if I ask you to draw a triangle with 2 lines and a circle, it’s not possible. You would need an additional line, thereby increasing the dimensionality.
    - If instead of adding a line, I add a star, you now have 5 dimensions. The key idea is that what really matters are the essential elements (3 lines for the triangle), but new, unnecessary elements are being added, which increases complexity.

- **Dimensionality in Datasets:**
    - Perhaps this example of adding unnecessary elements doesn’t seem intuitive. Let’s consider a dataset from a retail company. To add a simple product to the store, you might need to fill in 30 or 40 dimensions.
    - In this case, not all dimensions will be necessary, but they are still there. The machine needs to find patterns within these dimensions, which can seem advantageous at first because more data is available.
    - However, the more data you have, the more dispersed it can become, leading to a lack of symmetry. This makes it harder for the machine to understand and classify or predict accurately.

- **Impact on the Assignment:**
    - This is exactly what happened in the assignment. For example, several dimensions tracked how many times certain words appeared in emails, and whether they were spam or not. However, these dimensions had large differences, and they alone didn’t define whether an email was spam.
    - This dispersion of data reduces the effectiveness of the algorithm, making it harder to find a meaningful pattern.

- **Using PCA:**
    - To address this issue, **Principal Component Analysis (PCA)** was used. PCA allows you to reduce the number of dimensions without losing significant information. This technique helps to retain the integrity of the values while simplifying the dataset and improving the algorithm’s ability to find patterns.
